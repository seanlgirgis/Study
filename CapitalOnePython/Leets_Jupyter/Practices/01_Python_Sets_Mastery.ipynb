# Python Sets: The O(1) Superpower âš¡

**Why Sets Matter for Capital One / CodeSignal:**
- **Instant Lookup**: Checking `if x in seen` is O(1) on average. Lists are O(n).
- **Uniqueness**: Automatically removes duplicates.
- **Math Operations**: Easy intersection/union for "common elements" problems.

## 1. Creation & Basic Properties

```python
# 1. Creating a set
s1 = {9,1, 2, 3,1,7}
print(type(s1))
s2 = set([1, 2, 2, 3, 3, 3])  # From a list (removes duplicates!)

print(f"s1: {s1}")
print(f"s2: {s2} (Notice duplicates are gone)")

# 2. Empty set IS NOT {}
empty_dict = {}      # This is a DICT
empty_set = set()    # This is a SET

print(f"Type of {{}}: {type(empty_dict)}")
print(f"Type of set(): {type(empty_set)}")
```

## 2. Adding & Removing Elements

- `add(x)`: Adds x.
- `remove(x)`: Removes x, raises KeyError if missing.
- `discard(x)`: Removes x, DOES NOT raise error if missing (Safer!).

```python
my_set = {1, 2, 3}

my_set.add(4)
print(f"After add(4): {my_set}")

my_set.discard(2)
print(f"After discard(2): {my_set}")

my_set.discard(99)  # Safe! No error.
print(f"After discard(99): {my_set}")

# my_set.remove(99) # This would crash with KeyError
```

## 3. Set Math (Venn Diagrams in Code)

Very useful for "elements in both arrays" or "elements unique to array A".

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

# Union (All elements from both)
print(f"Union (a | b): {a | b}")

# Intersection (Common elements)
print(f"Intersection (a & b): {a & b}")

# Difference (In A but NOT in B)
print(f"Difference (a - b): {a - b}")

# Symmetric Difference (In A or B, but NOT both)
print(f"Symmetric Diff (a ^ b): {a ^ b}")
```

## 4. Performance: List vs Set Speed Test ðŸŽï¸

Let's simulate a large dataset search.

```python
import time

large_n = 10_000_000
large_list = list(range(large_n))
large_set = set(large_list)

target = large_n - 1  # Worst case for list (at the end)

# List Search
start = time.time()
found = target in large_list
list_time = time.time() - start
print(f"List search time: {list_time:.6f} seconds")

# Set Search
start = time.time()
found = target in large_set
set_time = time.time() - start
print(f"Set search time:  {set_time:.6f} seconds")

if set_time > 0:
    print(f"Sets are ~{list_time / set_time:.0f}x faster here!")
```

## 5. Practical Use Case: "Have I seen this before?"

Common pattern for `Two Sum`, `Contains Duplicate`, or `Sliding Window`.

```python
def has_duplicate(nums):
    seen = set()
    for num in nums:
        if num in seen:
            return True
        seen.add(num)
    return False

print(has_duplicate([1, 2, 3]))    # False
print(has_duplicate([1, 2, 3, 1])) # True
```

## 6. Myth Buster: "Are Sets Sorted?" ðŸ›‘

**Short Answer: NO.**

You might see `{1, 2, 3}` print as `{1, 2, 3}` and think it's sorted. That's just a coincidence for small integers (their hash is often the integer itself).

**Proof (Why NOT to trust set order):**

```python
# A set of strings or large numbers will NOT be sorted
s = {"apple", "banana", "cherry", "dog", "elephant"}
print(f"Set (Random Order): {s}")

# Order changes between runs or Python versions!
```

## 7. How to Enforce Order? (The Fix)

1. **Preserve Insertion Order** (1st added is 1st returned): **Use a Dictionary (Python 3.7+).** Keys are ordered by insertion.
2. **Sort by Value** (1, 2, 3...): Use `sorted()` to get a **List**.

```python
# Scenario 1: Keep Insertion Order (Unique list)
ordered_set = {}
ordered_set["first"] = True
ordered_set["second"] = True
ordered_set["third"] = True

print(f"Insertion Order (Dict Keys): {list(ordered_set.keys())}") # ['first', 'second', 'third']

# Scenario 2: Sorted Order (Value-based)
unsorted_set = {3, 1, 4, 1, 5, 9, 2}
sorted_list = sorted(unsorted_set)

print(f"Unsorted Set:   {unsorted_set}")
print(f"Sorted Result:  {sorted_list}")
```

## 8. Can I FORCE Randomness? (The "Chaos" Wrapper) ðŸŒ€

You can't easily change how Python handles `int` (hash(1) is always 1). However, you **can** wrap integers in a class to override their hash.
This **proves** sets rely on hash values, not sorting.


```python
import random

class ChaosInt:
    def __init__(self, val):
        self.val = val
        
    def __repr__(self):
        return str(self.val)
        
    def __hash__(self):
        # Return a RANDOM hash every time we call it (only for demo purposes!)
        # In real code, hash() must be stable for an object's lifetime.
        return random.randint(0, 1000000)
        
    def __eq__(self, other):
        return self.val == other.val

# Create a set of ChaosInts {1, 2, 3, 4, 5}
# Because their hashes are random, their positions in the set's internal table are random.
# Thus, the iteration order will be random.
chaos_set = {ChaosInt(1), ChaosInt(2), ChaosInt(3), ChaosInt(4), ChaosInt(5)}
print(f"Chaos Set: {chaos_set}")
# Run this cell multiple times -> The order will likely change!
```
