### LeetCode 56. Merge Intervals

Given an array of `intervals` where `intervals[i] = [starti, endi]`, merge all overlapping intervals, and return *an array of the non-overlapping intervals that cover all the intervals in the input*.

**Example 1:**
```
Input: intervals = [[1,3],[2,6],[8,10],[15,18]]
Output: [[1,6],[8,10],[15,18]]
Explanation: Since intervals [1,3] and [2,6] overlap, merge them into [1,6].
```

**Example 2:**
```
Input: intervals = [[1,4],[4,5]]
Output: [[1,5]]
Explanation: Intervals [1,4] and [4,5] are considered overlapping.
```

**Constraints:**
- `1 <= intervals.length <= 10^4`
- `intervals[i].length == 2`
- `0 <= starti <= endi <= 10^4`


````python
# ------------------------------------------------------------------
# Test Harness
# ------------------------------------------------------------------
def run_tests(func):
    test_cases = [
        ("Empty input", [], []),
        # --- LeetCode examples ---
        ("Example 1", [[1,3],[2,6],[8,10],[15,18]], [[1,6],[8,10],[15,18]]),
        ("Example 2", [[1,4],[4,5]], [[1,5]]),

        # --- Edge cases ---
        
        ("Single interval", [[1,4]], [[1,4]]),
        ("Nested contained", [[1,4],[2,3]], [[1,4]]),
        ("Already sorted, no overlaps", [[1,2],[3,4]], [[1,2],[3,4]]),
        ("Reverse order input", [[15,18],[8,10],[2,6],[1,3]], [[1,6],[8,10],[15,18]]),
        ("Touching at boundaries", [[1,4],[4,5]], [[1,5]]),
        ("Same intervals", [[1,4],[1,4]], [[1,4]]),
        ("All overlapping into one", [[1,4],[2,5],[3,6]], [[1,6]]),
        ("Empty input", [], []),
    ]

    print(f"Running tests for: {func.__name__}\n")
    passed = 0
    for desc, intervals, expected in test_cases:
        result = func(intervals)
        # Sorting the result just in case the order isn't guaranteed (though standard problem implies sorted output)
        result.sort()
        expected.sort()
        if result == expected:
            print(f"✅ {desc}: Passed")
            passed += 1
        else:
            print(f"❌ {desc}: Failed")
            print(f"   Input: {intervals}")
            print(f"   Expected: {expected}")
            print(f"   Actual:   {result}")
        print("-" * 30)
    print(f"\n{passed}/{len(test_cases)} passed\n")
````

````python
from typing import List

def merge(intervals: List[List[int]]) -> List[List[int]]:
    """
    LeetCode 56. Merge Intervals
    
    Given an array of intervals where intervals[i] = [starti, endi], 
    merge all overlapping intervals.
    To merge overlapping intervals we need to sort the intervals by item[0]  and then item[1]
    This will give us the first to the left interval
    """
    #ret: List[List[int]] = List()
    intervals.sort(key= lambda x: (x[0],x[1]))
    
    merged = []
    if not intervals:
        return []
    merged = [intervals[0]]
    for current in intervals[1:]:
        last = merged[-1]
        if current[0] <= last[1]: #Overlapping
            last[1] = max(last[1], current[1])
        else:
            merged.append(current)

    return merged
run_tests(merge)    
````

````python
from typing import List

def merge(intervals: List[List[int]]) -> List[List[int]]:
    """
    LeetCode 56. Merge Intervals
    
    Given an array of intervals where intervals[i] = [starti, endi], 
    merge all overlapping intervals.
    """
    if not intervals:
        return []
    
    # Sort intervals by start time
    intervals.sort(key=lambda x: x[0])
    
    merged = []
    for interval in intervals:
        # If list is empty or no overlap with previous merged interval, append it
        if not merged or merged[-1][1] < interval[0]:
            merged.append(interval)
        else:
            # Overlap exists, merge current into previous
            merged[-1][1] = max(merged[-1][1], interval[1])
            
    return merged

run_tests(merge)
````
