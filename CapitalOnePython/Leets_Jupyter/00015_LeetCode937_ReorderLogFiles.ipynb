### LeetCode 937. Reorder Data in Log Files

You are given an array of `logs`. Each log is a space-delimited string of words, where the first word is the **identifier**.

There are two types of logs:
1. **Letter-logs**: All words (except the identifier) consist of lowercase English letters.
2. **Digit-logs**: All words (except the identifier) consist of digits.

Reorder these logs so that:
1. The **letter-logs** come before all **digit-logs**.
2. The **letter-logs** are sorted lexicographically by their contents. If their contents are the same, then sort them lexicographically by their identifiers.
3. The **digit-logs** maintain their relative ordering.

Return *the final order of the logs*.

**Example 1:**
```
Input: logs = ["dig1 8 1 5 1","let1 art can","dig2 3 6","let2 own kit dig","let3 art zero"]
Output: ["let1 art can","let3 art zero","let2 own kit dig","dig1 8 1 5 1","dig2 3 6"]
Explanation:
The letter-log contents are all different, so their ordering is "art can", "art zero", "own kit dig".
The digit-logs have a relative order of "dig1 8 1 5 1", "dig2 3 6".
```

**Example 2:**
```
Input: logs = ["a1 9 2 3 1","g1 act car","zo4 4 7","ab1 off key dog","a8 act zoo"]
Output: ["g1 act car","a8 act zoo","ab1 off key dog","a1 9 2 3 1","zo4 4 7"]
```

**Constraints:**
- 1 ≤ logs.length ≤ 100
- 3 ≤ logs[i].length ≤ 100
- All the tokens of `logs[i]` are separated by a single space.
- `logs[i]` is guaranteed to have an identifier and at least one word after the identifier.


````python
from typing import List

def reorderLogFiles(logs: List[str]) -> List[str]:
    """
    LeetCode 937. Reorder Data in Log Files
    Reorders logs such that letter_logs come before digit_logss, sorted by content then id.
    Digit-logs maintain relative order.
    """
    letter_logs: List[str] = []
    digit_logs : List[str] = []
    for log in logs:
        if log.split(" " ,1)[1][0].isdigit(): digit_logs.append(log)
        else: letter_logs.append(log)
    letter_logs.sort(key= lambda log: (  log.split(" " ,1)[1] ,log.split(" " ,1)[0]  ))
    return letter_logs+digit_logs

# ------------------------------------------------------------------
# Test Harness
# ------------------------------------------------------------------
test_cases = [
    ("Example 1", 
     ["dig1 8 1 5 1","let1 art can","dig2 3 6","let2 own kit dig","let3 art zero"], 
     ["let1 art can","let3 art zero","let2 own kit dig","dig1 8 1 5 1","dig2 3 6"]),
    
    ("Example 2", 
     ["a1 9 2 3 1","g1 act car","zo4 4 7","ab1 off key dog","a8 act zoo"], 
     ["g1 act car","a8 act zoo","ab1 off key dog","a1 9 2 3 1","zo4 4 7"]),
    
    ("Only digit logs", 
     ["dig1 8 1 5 1", "dig2 3 6"], 
     ["dig1 8 1 5 1", "dig2 3 6"]),
    
    ("Only letter logs", 
     ["let2 own kit dig", "let1 art can"], 
     ["let1 art can", "let2 own kit dig"]),
    
    ("Tiebreaker: Same content, different ID", 
     ["id2 apple", "id1 apple"], 
     ["id1 apple", "id2 apple"]),
    
    ("Already sorted mixed", 
     ["a a", "b b", "c 1", "d 2"], 
     ["a a", "b b", "c 1", "d 2"]),
]

print("Running tests...\n")
for desc, logs, expected in test_cases:
    result = reorderLogFiles(logs[:])
    if result == expected:
        print(f"✅ {desc}: Passed")
    else:
        print(f"❌ {desc}: Failed")
        print(f"   Input: {logs}")
        print(f"   Expected: {expected}")
        print(f"   Actual:   {result}")
    print("-" * 30)
````
